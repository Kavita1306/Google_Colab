```python
import matplotlib.pyplot as plt
fig,gnt=plt.subplots()
gnt.set_ylim(0,50)
gnt.set_xlim(0,100)
gnt.set_xlabel('x-axis')
gnt.set_ylabel('y-axis')
gnt.set_yticks([10,25,30])
gnt.set_ytickslabels(['1','2','3'])
gnt.grid(True)
gnt.broken_barh([20,40)],(30,8),forcolors=('tab:green'))
gnt.broken_barh([50,30)],(60,7),(10,5),forcolors=('tab:blue'))
gnt.broken_barh([10,50)],(90,9),(100,50),(10,7),forcolors=('tab:orange'))

```
